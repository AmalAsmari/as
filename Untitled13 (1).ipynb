```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('aaa')
```

```python
df
```

```python
df.plot(kind='box', figsize=(12, 6))


plt.show()
```

```python
df['Price'].plot(kind='pie',
                            figsize=(5, 6),
                            autopct='%1.1f%%', 
                            startangle=90,     
                            shadow=True,            
                            )
plt.axis('equal')
plt.show()
```
